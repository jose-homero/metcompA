```python
import numpy as np #importação biblioteca numpy
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt #importação pyplot para gráficos
def posição(t,vt,tau):
  y= (-vt)*t+(-vt*tau)*(np.exp(-t/tau))+250
  return y
def velocidade (t,vt,tau):
  v=(-vt)+vt*(np.exp(-t/tau))
  return v
def aceleração(t,vt,tau):
  a=(-vt)*(np.exp(-t/tau)*(1/tau))
  return a

t=np.arange(0.,10.5,0.5)

pos=posição(t,vt=20,tau=5)# array posição

vel=velocidade(t,vt=20,tau=5)#'' velocidade

ace=aceleração(t,vt=20,tau=5) #'' aceleração

output=open('dados_queda.txt', 'w') #abre arquivo
output.write(f'#t pos vel\n') # escrita no arquivo

for i in range(len(t)):
  output.write(f'{t[i]:.2f} {pos[i]:.3f} {vel[i]:.3f} {ace[i]:.3f}\n')
output.close()

with PdfPages('queda.pdf') as pdf:
  figura = plt.figure(figsize=(10, 5))
  plt.plot(t,pos, label='posição x tempo', color='green')
  plt.xlabel('tempo(s)')
  plt.ylabel('altura(m)')
  plt.legend()
  pdf.savefig()
  plt.show()

  figura = plt.figure(figsize=(9.8, 5))
  plt.plot(t,vel, label='velocidade x tempo', color='purple')
  plt.xlabel('tempo(s)')
  plt.ylabel('velocidade(m/s)')
  plt.legend()
  pdf.savefig()
  plt.show()

  figura = plt.figure(figsize=(9.9, 5))
  plt.plot(t,ace, label='aceleração x tempo', color='red')
  plt.xlabel('tempo(s)')
  plt.ylabel('aceleração(m/s²)')
  plt.legend()
  pdf.savefig()
  plt.show()
```
